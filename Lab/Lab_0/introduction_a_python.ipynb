## Variables en Python

Une **variable** est un nom qui permet de stocker une valeur en mémoire. On peut ensuite réutiliser cette valeur dans des calculs ou l’afficher.

En Python, on crée une variable simplement en lui attribuant une valeur avec le signe ``=``

## Les structures de données

Python supporte nativement plusieurs structures de données importantes :

Elles permettent de stocker et manipuler des collections de valeurs.

### Les listes

Les listes sont une structure de données très souple :

* elles peuvent contenir des valeurs de différents types (entiers, chaînes, flottants, …),
* on peut ajouter, enlever, réordonner et modifier leurs éléments,
* elles sont créées à l’aide de crochets `[]`.

```python
l = []  # Initialisation d'une liste vide
print(l)

l = [10, "Bonjour", 3.14, True]  # initialisation d'une liste non vide
print(l)

l.append(7)  # ajout de l'entier 7 à la fin de la liste
print(l)

print(l[1]) # afficher le deuxième élément

l[0] = 5 # modifier le premier élément
print(l)
```

Il est possible de générer des listes par *compréhension*

```python
l = [i ** 2 for i in range(10)]  # liste des carrés de 0 à 9 compris
print(l)
```

```python
l = [i**2 for i in range(10) if i%2] # liste des carrés des impaires de 0 à 9 compris 
print(l)
```

On peut obtenir la longueur d'une liste avec la fonction `len` qui renvoie un entier.
Cette fonction peut être utilisée pour toutes les structures de données présentées ici.

```python
print(len(l))
```

### Les tuples

Les **tuples** sont une structure de données très proche des listes, mais :

* ils utilisent des parenthèses () au lieu des crochets [],
* ils sont **immuables** : une fois créés, on **ne peut pas modifier** leurs éléments (ajouter, enlever ou réordonner).

Les tuples sont donc utiles lorsque l’on veut regrouper des données **fixes** qui ne doivent pas être modifiées.

```python
t = (1, 'deux', 3.5, False)
print(t)

t[2] = 4  # renvoie une erreur parce qu'un tuple est immuable
```

### L'indexage et le découpage (Slicing)

Les **listes** et les **tuples** peuvent être indiciés (accès à un élément précis) et **découpés** (slicing, sélection d’une portion de la structure).

* Indexage : permet de récupérer un élément par son indice (attention, l’indice commence à `0`).
* Découpage (slicing) : permet d’extraire une sous-partie en utilisant la notation `[début:fin:pas]`.
  * `début` : indice de départ (inclus, par défaut `0`)
  * `fin` : indice de fin (**exclu**)
  * `pas` : intervalle entre les éléments (par défaut `1`)

```python
l = [i * 10 for i in range(10)]
print("la liste l : {}".format(l))
print(l[1])  # sélectionne et affiche le deuxième élément
print(l[:3])  # sélectionne et affiche les trois premiers éléments
print(l[3:])  # sélectionne et affiche les éléments de liste à partir du quatrième
print(l[-1])  # sélectionne et affiche le dernier élément
print(l[-3:])  # sélectionne et affiche les trois derniers éléments
print(l[1:-1:2])  # sélectionne et affiche : du deuxième à l'avant dernier avec un pas de 2
```

### Les ensembles

Un **ensemble** (`set`) est une collection **non ordonnée** d’éléments **uniques** (pas de doublons).

* Délimité par des accolades `{}`
* Ou créé avec la fonction `set()`

> Attention :
> * `{}` crée un **dictionnaire vide**.
> * Pour créer un **ensemble vide**, il faut utiliser `set()`.

Les ensembles permettent d'effectuer des opérations mathématiques telles que les unions, intersections, différences, différences symétriques, etc.

```python
s1 = set([1, 2, 3, 4])
s2 = {2, 2, 1}

print(s1)
print(len(s1))

print(s2) # Les ensembles ne gardent qu’un seul exemplaire d’un élément
print(len(s2))
```

Il est pratique d'utiliser les ensembles pour supprimer les doublons dans une liste

```python
l = [1,1,2,3,4,5,6,7,6,8,9,8,9]
print(l)
print(list(set(l)))
```

Les ensembles permettent d’effectuer des opérations mathématiques classiques :

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A | B)   # Union
print(A & B)   # Intersection
print(A - B)   # Différence
print(A ^ B)   # Différence symétrique
```

### Les dictionnaires

Les **dictionnaires** sont une structure de données **associative** : chaque élément est composé d’une **clé** et d’une **valeur**.

* Les **clés** doivent être **uniques** et **immuables** (ex. : entiers, chaînes, tuples).
* Les **valeurs** peuvent être de n’importe quel type (nombres, chaînes, listes, autres dictionnaires…).
* On les délimite par des accolades `{}`, et l’association se fait avec un double point `:` (`clé: valeur`).

```python
etudiant = {"nom": "Alice", "age": 23, "programme": "Génie", ("ville", "pays"): ("Montréal", "Canada")}

print(etudiant)
print(etudiant["nom"])
print(etudiant["programme"])

etudiant["age"] = 24              # modification
etudiant["université"] = "ÉTS"    # ajout
print(etudiant)
```

Pour obtenir les clés du dictionnaire, on utilise la méthode `keys`:

```python
print(etudiant.keys())     # liste des clés
```

Pour obtenir les valeurs du dictionnaire, on utilise la méthode `values`:

```python
print(etudiant.values())   # liste des valeurs
```

### Les conditions : `if`, `elif` et `else`

Comme dans la plupart des langages de programmation, Python permet d’exécuter des instructions **seulement si une condition est vraie**.
On utilise pour cela les mots-clés :

* `if` → si la condition est vraie, le bloc associé est exécuté.
* `elif` (else if) → permet de tester une autre condition si la précédente est fausse.
* `else` → bloc exécuté uniquement si toutes les conditions précédentes sont fausses.

> En Python, l’indentation (les espaces au début de la ligne) est **obligatoire** pour définir les blocs de code.

```python
x = 7

if x > 10:
    print("x est supérieur à 10")
elif x == 10:
    print("x est exactement 10")
else:
    print("x est inférieur à 10")
```

* `not` → inverse la valeur d’une condition

### Exercice : Filtrage des rendez-vous médicaux selon les symptômes

On veut écrire un petit programme qui aide à trier les patients en fonction de leurs symptômes afin de décider de l'urgence du rendez-vous.

Règles:

* Rendez-vous urgent : `"douleur thoracique"` ou `"essoufflement"`.

* Rendez-vous rapide : `"fièvre"` ou `"toux"`.

* Rendez-vous normal : `"maux de tête"`.

---

```python
patients_list = []
urgent = []
rapide = []
normal = []

def add_patient(name, age, symptome):
    patient = {
        "name": name,
        "age": age,
        "symptome": symptome
    }
    patients_list.append(patient)
    print(f"Patient {name} ajouté.")

def tri_patients_by_urgency():
    for patient in patients_list:
        symptome = patient['symptome']
        if symptome in ["douleur thoracique", "essoufflement"]:
            urgent.append(patient)
        elif symptome in ["fièvre", "toux"]:
            rapide.append(patient)
        elif symptome == "maux de tête":
            normal.append(patient)
    
    print("Rendez-vous urgent :")
    for p in urgent:
        print(f"Nom: {p['name']}, Âge: {p['age']}, Symptôme: {p['symptome']}")
    print("\nRendez-vous rapide :")
    for p in rapide:
        print(f"Nom: {p['name']}, Âge: {p['age']}, Symptôme: {p['symptome']}")
    print("\nRendez-vous normal :")
    for p in normal:
        print(f"Nom: {p['name']}, Âge: {p['age']}, Symptôme: {p['symptome']}")
    return urgent, rapide, normal

# Exemple d'utilisation
add_patient("Alice", 30, "douleur thoracique")
add_patient("Bob", 45, "fièvre")
add_patient("Charlie", 25, "maux de tête")
add_patient("David", 50, "essoufflement")
add_patient("Emma", 22, "toux")
print("============\n")
urgent, rapide, normal = tri_patients_by_urgency()
```

### Les boucles

En Python, il existe deux types principaux de boucles :

* `while` : répète des instructions tant qu’une condition est vraie.
* `for` : répète des instructions en parcourant une séquence (liste, tuple, chaîne, etc.).

#### Boucle `while`

La boucle while continue d’exécuter un bloc d’instructions tant que la condition est **vraie**.

> Attention : si la condition n’est jamais fausse, on obtient une **boucle infinie**.

```python
compteur = 0

while compteur < 5:
    print("Compteur =", compteur)
    compteur += 1   # équivaut à compteur = compteur + 1
```

#### Boucle `for`

La boucle for permet d’**itérer** (parcourir) directement une **collection de valeurs** (liste, tuple, dictionnaire, chaîne de caractères, etc.).

Contrairement à certains autres langages (comme C ou Java), en Python la boucle `for` ne repose pas sur des indices numériques, mais sur les **éléments eux-mêmes**.

```python
fruits = ["Pomme", "Banane", "Orange"]

for fruit in fruits:
    print(fruit)
```

La fonction `range()` permet de créer une **séquence de nombres entiers** qui peut être parcourue dans une boucle `for`.
Par défaut, `range(n)` génère les entiers de **0 jusqu’à n-1**.

```python
for i in range(5):
    print(i)
```

On dispose d’une liste de mesures de la fréquence cardiaque (en battements par minute, bpm).

[72, 55, 110, 95, 130, 60, 45]

On veut écrire un programme qui :

Affiche chaque valeur avec un message :
* `"Tachycardie"` si la fréquence > 100 bpm
* `"Bradycardie"` si la fréquence < 60 bpm
* `"Normale"` sinon

Compte combien de mesures sont `"anormales"` (tachycardie ou bradycardie).

---

```python
frequence_cardiaque = [72, 55, 110, 95, 130, 60, 45]
counter = 0

for i in frequence_cardiaque:
    if i > 100:
        print(f"Tachycardie - {i} bpm")
        counter += 1
    elif i < 60:
       print(f"Brachycardie - {i} bpm")
       counter += 1
    else:
        print(f"Normale - {i} bpm")
print(f"Nombre d'anomalies détectées : {counter}")
```

### Exercice : Fonction calculatrice simple

Écrire une fonction calc(sign, x, y) qui effectue une opération mathématique selon le signe donné :
* `"+"` → addition
* `"-"` → soustraction
* `"*"` → multiplication
* `"/"` → division

La fonction doit retourner le résultat de l’opération.

---

```python
def calc(sign, x, y):
    if sign == '+':
        return x + y
    elif sign == '-':
        return x - y
    elif sign == '*':
        return x * y
    elif sign == '/':
        if y != 0:
            return x / y
        else:
            return "Erreur : Division par zéro"
    else:
        return "Erreur : Opérateur non reconnu"
    
print(calc('+', 3, 5))
print(calc('-', 10, 4))
print(calc('*', 2, 6))
print(calc('/', 8, 2))
print(calc('/', 8, 0))
```

## Les classes

Python est un langage **orienté objet**.
* On peut définir des **classes** en Python.
* Avec les classes, on peut **modéliser presque n’importe quoi**.
* Par convention, les noms de classes suivent le format `NomDeClasse`.

À l’intérieur d’une classe, on définit des **fonctions associées à la classe**, appelées **méthodes**.

La méthode la plus courante est `__init__(self)`, appelée **constructeur** :
* Elle sert à créer des **instances de la classe** avec un état initial défini.

Exemple simple :

```python
class Patient:
    # Le constructeur __init__ initialise l'état de l'objet
    def __init__(self, nom, age, frequence_cardiaque):
        # self fait référence à l'instance créée
        self.nom = nom
        self.age = age
        self.frequence_cardiaque = frequence_cardiaque

    # Méthode pour afficher les informations du patient
    def afficher_info(self):
        print(f"Nom : {self.nom}, Âge : {self.age} ans, Fréquence cardiaque : {self.frequence_cardiaque} bpm")


# Création de deux instances de la classe Patient
p1 = Patient("Alice", 35, 72)
p2 = Patient("Bob", 60, 110)

print(p1.nom)
print(p2.age)

# Appel d'une méthode
p1.afficher_info()
p2.afficher_info()
```

### Lecture et écriture de fichier

En Python, la lecture et l’écriture de fichiers texte sont très simples.
On utilise la fonction `open()` qui prend deux arguments principaux :

* le **nom du fichier** (par ex. `"test.txt"`)
* le **mode d’ouverture** (`'w'`, `'r'`, `'a'`, …)

**Modes courants**

* `'w'` → écriture (**write**) : crée un nouveau fichier (ou écrase s’il existe déjà)
* `'r'` → lecture (**read**) : ouvre un fichier existant en lecture seule
* `'a'` → ajout (**append**) : ajoute du texte à la fin du fichier

```python
l1 = 'Ceci est la première ligne\n'  # \n = retour à la ligne
l2 = 'Ceci est la deuxième ligne'

f = open('test.txt', 'w')  # ouverture en écriture
f.write(l1)
f.write(l2)
f.close()  # fermeture du fichier
```

Il est recommandé de ne pas utiliser la syntaxe précédente avec `open()` et `close()`, car on peut oublier de fermer le fichier ou rencontrer des erreurs si une exception survient avant la fermeture.

La **bonne pratique** consiste à utiliser la syntaxe `with ... as`, qui garantit que le fichier sera automatiquement fermé après les opérations, même en cas d’erreur.

Le code précédent devient :

```python
with open('test.txt', 'w') as f:
    f.write(l1)
    f.write(l2)
```

De façon similaire pour la lecture:

```python
with open('test.txt', 'r') as f:  # option 'r' pour lire dans un fichier
    text = f.read()
print(text)
```

Quand on lit un fichier entier avec `read()`, on obtient **une seule chaîne de caractères** qui contient tout le contenu du fichier.

Si on veut retrouver les lignes séparées, on peut utiliser la méthode `.splitlines()`, qui découpe automatiquement la chaîne en fonction des retours à la ligne.

```python
lines = text.splitlines()
print(lines)
```

## NumPy

**NumPy** est un paquet fondamental en Python pour les calculs numériques et la manipulation de **tableaux multidimensionnels**.

Toutes les opérations que l’on fait avec NumPy pourraient être faites avec les structures de base de Python (comme les listes), mais :
* elles seraient **beaucoup plus lentes**,
* et le code serait **plus lourd à écrire**.

Un **tableau** est une structure de données représentant une séquence finie d’éléments.
* On y accède efficacement grâce à l’**indice**.
* En Python, les listes peuvent être vues comme des **tableaux à une dimension**.

*En NumPy* :
* un tableau 1D correspond à un vecteur,
* un tableau 2D correspond à une matrice (séquence de lignes),
* on peut aussi créer des tableaux à plus de deux dimensions (tenseur).

Différences entre listes Python et tableaux NumPy
* **Taille fixe** : la taille d’un tableau NumPy est définie à sa création.
* **Type homogène** : tous les éléments d’un tableau NumPy doivent avoir le **même type** (int, float, etc.).

> Cela rend les tableaux NumPy moins flexibles que les listes Python.
Mais en contrepartie, ils sont **beaucoup plus rapides** et facilitent l’écriture de calculs complexes.

La documentation est excellente : https://numpy.org/doc/stable/index.html

### Création des tableaux

NumPy propose de nombreuses façons de créer des **tableaux** selon le cas d’usage.
La syntaxe est similaire : on précise la **taille du tableau** et éventuellement le **type de données** (par défaut `float64`).

Pour la liste complète, voir https://numpy.org/doc/stable/reference/routines.array-creation.html.

1. Tableaux vides ou initialisés
* `np.empty(shape)` → crée un tableau vide (valeurs indéterminées).
* `np.zeros(shape)` → rempli de 0.
* `np.ones(shape)` → rempli de 1.
* `np.full(shape, value)` → rempli d’une valeur spécifique.

```python
import numpy as np

print(np.zeros((2, 3)))
print(np.ones((2, 3)))
print(np.full((2, 3), 7))
```

2. Vecteurs avec valeurs espacées
* `np.arange(start, stop, step)` → valeurs espacées d’un **pas** donné.
* `np.linspace(start, stop, num)` → valeurs espacées **régulièrement** (inclus `stop`).

```python
print(np.arange(0, 10, 2))
print(np.linspace(0, 1, 5)) 
```

3. Matrices identité et diagonales
* np.identity(n) → matrice identité carrée.
* np.eye(N, M) → matrice diagonale de taille N x M

```python
print(np.identity(3))
print(np.eye(3, 4))
```

4. À partir de listes ou listes imbriquées
* np.array(liste) → transforme une liste Python en tableau NumPy.

```python
print(np.array([1, 2, 3]))            # vecteur
print(np.array([[1, 2], [3, 4]]))     # matrice 2x2
```

5. Fonctions "like"
Créent un tableau de même **taille** et **type** qu’un tableau existant :
* `np.empty_like(a)`
* `np.zeros_like(a)`
* `np.ones_like(a)`
* `np.full_like(a, value)`

```python
a = np.array([[1, 2, 3], [4, 5, 6]])

print(np.zeros_like(a))   # même taille que a, rempli de 0
print(np.ones_like(a))    # même taille que a, rempli de 1
print(np.full_like(a, 9)) # même taille que a, rempli de 9
```

### Attributs des tableaux

Il est fréquent de vouloir obtenir des informations sur les **tableaux NumPy** que l’on manipule.
Pour cela, NumPy fournit plusieurs **attributs utiles** :
* `shape` → retourne la taille du tableau sous forme d’un **tuple**
* `ndim` → retourne le **nombre de dimensions** (équivalent à `len(shape)`)
* `dtype` → retourne le **type de données** des éléments
* `size` → retourne le **nombre total d’éléments** dans le tableau

```python
L = [[[17, 42], [-5.3, 0]]]
a = np.array(L)
print(f'a est un tableau à {a.ndim} dimensions, de taille {a.shape}, contenant {a.size} éléments de type {a.dtype}')
```

On peut modifier la **géométrie** (forme) d’un tableau avec la fonction `np.reshape()` ou la méthode `.reshape()` d’un tableau NumPy.

*Condition importante* :
* Le **nombre total d’éléments** doit rester le même.
* On peut mettre `-1` **pour une seule dimension** : NumPy calculera automatiquement la taille correcte de cette dimension.
* L’ordre des éléments ne change pas → NumPy lit *de gauche à droite, puis de haut en bas*.

```python
a = np.arange(12)    # tableau de 0 à 11
print("a :", a)

b = a.reshape((3, 4))   # 3 lignes, 4 colonnes
print("\nreshape (3,4) :\n", b)

c = a.reshape((6, -1))   # 6 lignes, 2 colonnes
print("\nreshape (6,2) :\n", c)
```

### Fonctions mathématiques de base dans NumPy

NumPy étant destiné au calcul scientifique, il propose de nombreuses **fonctions mathématiques de base** déjà implémentées :
* exponentielle : `np.exp`
* logarithmes : `np.log` (naturel, base e), `np.log10` (base 10), `np.log2` (base 2)
* trigonométrie : `np.sin`, `np.cos`, `np.tan`, etc.

*Attention* :
* `np.log(x)` correspond au **logarithme naturel** (ln(x)).
* Pour les logarithmes **décimal** ou **binaire**, on utilise respectivement `np.log10(x)` et `np.log2(x)` 

Pour une liste complète de ces fonctions, voir https://numpy.org/doc/stable/reference/routines.math.html.

```python
a = np.array([1, 2, 3, 4])

print("Exponentielle :", np.exp(a))
print("Log naturel   :", np.log(a))
print("Log base 10   :", np.log10(a))
print("Log base 2    :", np.log2(a))

angles = np.array([0, np.pi/2, np.pi])
print("sin :", np.sin(angles))
print("cos :", np.cos(angles))
```

#### Algèbre linéaire et produit matriciel

NumPy implémente la plupart des opérations d’**algèbre linéaire** (produits, transposée, inverses, déterminants, etc.).

Attention à la différence entre deux opérateurs :

* `*` → produit **élément par élément** (aussi appelé Hadamard product)
* `@` → produit **matriciel** (comme en algèbre linéaire)

Les nombreuses fonctions d'algèbre linéaire sont référencées sur cette page: https://numpy.org/doc/stable/reference/routines.linalg.html. 

```python
A = np.array([[1, 1], [2, 2]])
B = np.array([[-1, 1], [1, 1]])

print('A :', A, sep='\n')
print('B :', B, sep='\n')
print('Transposée de A :', A.T, sep='\n')  # .T est considéré comme un attribut
print('Produit matriciel AB :', A @ B, sep='\n')
print('Produit de Hadamard de A et B :', A * B, sep='\n')
print('Inverse de B :',np.linalg.inv(B),sep='\n')
```

### L'indexage

L’une des grandes forces de NumPy est sa capacité à **indexer efficacement les éléments** d’un tableau.
* NumPy supporte un indexage similaire à celui des listes Python,
* mais il offre aussi des fonctionnalités **beaucoup plus avancées** (slicing, masques booléens, indexation avancée).

Pour aller plus loin: https://numpy.org/doc/stable/user/basics.indexing.html#basics-indexing.

#### L'indexage d'élément unique

Comme pour les listes Python, on peut accéder à un élément en indiquant son **indice**.
*La différence* : pour les tableaux multidimensionnels, NumPy permet d’indiquer les indices **directement séparés par des virgules** au lieu d’utiliser plusieurs crochets imbriqués.

```python
A = np.arange(12).reshape((3, 4))
print(A)

print('Deuxième élément de la première ligne :', A[0, 1])
print('Quatrième élément de la deuxième ligne :', A[1, 3])  # retourne "7": l'élément de la ligne 1, colonne 2
```

#### Le découpage (slicing)

En NumPy, on peut utiliser le symbole `:` pour **extraire des sous-parties d’un tableau**.
* Utilisé seul, `:` signifie **“tous les éléments le long de cette dimension”**.
* Utilisé avec des **indices**, il permet de sélectionner une portion d’un tableau :
    * `[start:stop]` → éléments de `start` à `stop-1` (le stop est exclu)
    * `[start:stop:step]` → avec un pas (step)

```python
print('Première colonne :', A[:, 0])
print('Deuxième ligne :', A[1, :])  # équivalent à A[1]
print('Toutes les colonnes sauf la première :', A[:, 1:], sep='\n')
print('Deuxième et troisième colonnes des deux premières lignes :', A[:2, 1:3], sep='\n')
```

#### L'indexage avancé

Avec NumPy, on peut aller plus loin que l’indexation classique.
L’**indexage avancé** permet de sélectionner des éléments d’un tableau en fournissant :
* une **liste** ou un **tableau NumPy** d’indices,
* ce qui permet de récupérer **plusieurs positions spécifiques** en une seule opération

```python
print(A[[0, 2], [1, 3]])
print(A[np.array([1, 0]), np.array([2, -1])])
```

#### L'indexage booléen ou l'utilisation de masque

L’**indexage booléen** consiste à sélectionner des éléments d’un tableau en utilisant un **masque** (un tableau de booléens `True/False`).
* `True` → l’élément est **sélectionné**.
* `False` → l’élément est **ignoré**.

Le masque doit avoir la **même forme** que le tableau de départ.

```python
mask = np.array([[False,  True,  True, False],
                 [ True,  True, False, False],
                 [ True, False, False,  True]])
print(A[mask])
```

Ainsi, on a sélectionné seulement les valeurs pour lesquelles le tableau `mask` est vrai. Cela est très utile  pour filtrer des tableaux selon des critères arbitraires.

On peut aussi faire cet indexage seulement sur certaines dimensions.

```python
print('Tous les éléments impairs de A :', A[A % 2 != 0])
print('La deuxième et troisième colonne :', A[:, np.array([False, True, False, True])], sep='\n')
```

## Matplotlib

**Matplotlib** est un paquet très utilisé pour **créer et afficher des figures** en Python.
C’est l’outil de base pour la **visualisation scientifique** (graphiques, courbes, histogrammes, etc.).

En pratique, c’est le sous-module `pyplot` que l’on utilise le plus, et on l’importe généralement sous l’abréviation `plt` :

```python
import matplotlib.pyplot as plt
import numpy as np

# Données
x = np.linspace(0, 4*np.pi, 100)   # 100 points de 0 à 4π
print(x)
y = np.sin(x)
```

```python
# Tracé
plt.plot(x, y, label="sin(x)")
```

En soit, ceci en fait un outil très utile, facilitant grandement la création de visualisations. Il est aussi possible de personnaliser ces graphiques jusqu'au moindre détail avec l'API orientée objet, qu'il est recommandé d'utiliser :

```python
fig, ax = plt.subplots()
ax.plot(x, y, label='sin')

# ajoute des labels sur le graphique et les axes
ax.set_title('Fonctions trigonométriques')
ax.set_xlabel('t')
ax.set_ylabel('f(t)')
# change les limites du graphique
ax.set_xlim(x.min(), x.max())
# change les graduations et les labels des graduations
ax.set_xticks(np.linspace(start=0, stop=4 * np.pi, num=9))
ax.set_xticklabels(['0', r'$\dfrac{\pi}{2}$', r'$\pi$', r'$\dfrac{3\pi}{2}$', r'$2\pi$',
                    r'$\dfrac{5\pi}{2}$', r'$3\pi$', r'$\dfrac{7\pi}{2}$', r'$4\pi$'])
# ajoute une grille et la légende dans la graphique
ax.grid(linestyle='--', linewidth=1)
ax.legend()
```

## Pandas

**Pandas** est une bibliothèque Python qui facilite la manipulation et l’analyse des données.
Elle est particulièrement utile pour travailler avec des données **tabulaires** (comme des fichiers CSV, Excel, ou des bases de données).

La structure de base en pandas est le **DataFrame**, qui ressemble à un tableau Excel ou une table SQL :
* Chaque colonne peut avoir un type différent (nombres, chaînes de caractères, dates, etc.).
* Chaque ligne correspond à une observation ou un enregistrement.

La fonction la plus utilisée est `pd.read_csv`, qui permet de lire directement des fichiers **CSV** (Comma-Separated Values).

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("eval.csv")

# Vérifier le contenu
print(df.head())

# Accéder à une colonne
print(df["Pondération"])

# Calculer la somme des pondérations
print("Somme des pondérations :", df["Pondération"].sum())

# Extraire les colonnes utiles
labels = df["Activité"]
sizes = df["Pondération"]

# Créer le graphique en camembert
plt.pie(
    sizes, 
    labels=labels, 
    autopct='%1.1f%%',  # afficher les pourcentages
)

plt.title("Répartition des pondérations")
```

Pandas est beaucoup plus flexible que NumPy pour manipuler des fichiers réels contenant des colonnes de données hétérogènes.

C’est la bibliothèque de référence en Data Science et Machine Learning.
